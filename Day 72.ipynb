## Objective
The objective of this assignment is to learn the basic workflow of text classification using Natural Language Processing (NLP) techniques in Python. You will preprocess text data, extract features, build a machine learning model, and evaluate its performance.

```python
import warnings
import sys
if not sys.warnoptions:
    warnings.simplefilter("ignore")
```

### Task 1: Import Necessary Libraries

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

### Task 2: Create a Dataset

```python
# Sample dataset
texts = [
    "I love this product", 
    "This is terrible", 
    "Absolutely fantastic!", 
    "Not good at all", 
    "Worth every penny", 
    "Awful experience", 
    "Excellent quality", 
    "I hate it",
    "Superb performance",
    "Horrible design"
]
labels = ["positive", "negative", "positive", "negative", "positive", "negative", "positive", "negative", "positive", "negative"]
```

### Task 3: Preprocess the Data

```python
# Preprocessing and Feature Extraction
vectorizer = CountVectorizer()  # Initialize the vectorizer
X = vectorizer.fit_transform(texts)  # Transform text to BoW representation
```

### Task 4: Split the Data

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=42)
```

### Task 5: Train a Machine Learning Model

```python
# Train a Naive Bayes classifier
model = MultinomialNB()
model.fit(X_train, y_train)
```

### Task 6: Make Predictions

```python
# Make predictions
y_pred = model.predict(X_test)
```

### Task 7: Evaluate the Model

```python
# Evaluate model performance
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
